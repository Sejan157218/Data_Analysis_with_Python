```python
# reading line by line

# automacally close file

with open ('./titanic_test.csv') as file:
    file_content = file.readlines()
print(file_content)
```

```python
file_content[0].strip()
```

```python
# Csv to object of array

headers = file_content[0].strip().split(',')
print(headers)
```

```python
array = []

for items in file_content[1:]:
    obj = {}
    for head, value in zip(headers,items.strip().split(',')):
        obj[f"{head}"] = value

    array.append(obj)
print(array)
```

```python
# write csv

with open('./write_csv.csv', 'w') as f:
    heads = array[0].keys()
    f.write(','.join(heads)+ '\n')
    
    for items in array:
        values =[]
        for head in heads:
            values.append(str(items.get(head,"")))
        f.write(','.join(values)+ '\n')
```
